```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the annual average temperature data
annual_df = pd.read_csv('data.csv', skiprows=4)  # Adjust the filename as needed
annual_df.columns = ['Year', 'Value']

# Prepare the data for regression
X = annual_df['Year'].values.reshape(-1, 1)
y = annual_df['Value'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(annual_df['Year'], y, label='Actual')
plt.plot(annual_df['Year'], y_pred, color='red', label='Trend Line')
plt.title('Average Annual Temperature in NYC (1895-2025)')
plt.xlabel('Year')
plt.ylabel('Temperature (F)')
plt.legend()
plt.grid(True)
plt.show()
```

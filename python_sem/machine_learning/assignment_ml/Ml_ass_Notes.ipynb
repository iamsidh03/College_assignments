```python
#Q1
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
iris=load_iris()
x=iris.data
y=iris.target
labels=iris.target_names
tsne=TSNE(n_components=2,random_state=42)
x_2d=tsne.fit_transform(x)
plt.figure(figsize=(8,6))
for i in range(len(labels)):
    plt.scatter(x_2d[y==i,0],x_2d[y==i,1],label=labels[i])
plt.title('t-SNE on iris Dataset')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q2
# Step 1: Import necessary libraries
import pandas as pd                     # For working with data in table format
import seaborn as sns                   # For creating beautiful plots easily
import matplotlib.pyplot as plt         # For displaying the plot
from sklearn.datasets import fetch_california_housing  # To load the dataset

# Step 2: Load the California Housing dataset
housing = fetch_california_housing()    # Loads the data as a dictionary-like object

# Step 3: Convert it into a pandas DataFrame
df = pd.DataFrame(housing.data, columns=housing.feature_names)

# Step 4: Add the target (housing prices) to the DataFrame
df['Target'] = housing.target

# Step 5: Create a pairplot using seaborn
# WARNING: Pairplot plots all combinations, so we select only a few columns for speed and clarity
selected_columns = ['MedInc', 'AveRooms', 'AveOccup', 'Target']
sns.pairplot(df[selected_columns], diag_kind='kde')  # kde = kernel density estimate for smoother histograms

# Step 6: Show the plot with interactive toolbar
plt.suptitle("California Housing Data - Pairplot", y=1.02)
plt.tight_layout()
plt.show()
```

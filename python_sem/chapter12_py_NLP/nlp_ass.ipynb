```python
# 5
# Step 1: Import necessary libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Step 2: Download the required resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Step 3: Sample text
text = "Natural Language Processing enables machines to understand and process human languages."

# Step 4: Tokenize the text into words
words = word_tokenize(text)

# Step 5: Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = []
for word in words:
    if word.lower() not in stop_words and word.isalpha():
        filtered_words.append(word)

# Step 6: Stemming
stemmer = PorterStemmer()
stemmed_words = []
for word in filtered_words:
    stemmed_words.append(stemmer.stem(word))

# Step 7: Lemmatization
    lemmatizer = WordNetLemmatizer()
lemmatized_words = []
for word in filtered_words:
    lemmatized_words.append(lemmatizer.lemmatize(word))

# Step 8: Print results
print("Original Words:", words)
print("After Stopword Removal:", filtered_words)
print("After Stemming:", stemmed_words)
print("After Lemmatization:", lemmatized_words)
```

```python
# Step 1: Import necessary libraries
import requests
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# requests: download the webpage
# BeautifulSoup:extract readable text from HTML
# worldcloud: create the word cloud image
# matplotlib.pyplot:diaplay the image


# Step 2: Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Step 3: Download the webpage
url = "https://www.python.org"
response = requests.get(url)

# Step 4: Parse the HTML with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
# Parses the HTML code using BeautifulSoup so we can extract the text easily

# Step 5: Extract text from the page
text = soup.get_text()#remove all  HTML tags

# Step 6: Tokenize and remove stopwords
words = word_tokenize(text)
stop_words = set(stopwords.words('english'))
# stopwords.words('english') gives you common stopwords in english language that doesnot add much meaning 

filtered_words = [word for word in words if word.isalpha() and word.lower() not in stop_words]
cleaned_text = ' '.join(filtered_words)#Join all the words in the list using a space between them."

# Step 7: Create and display the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cleaned_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#7
from textblob import TextBlob
text = """Natural Language Processing enables machines to understand and process human languages.
It is a fascinating field with numerous applications, such as chatbots and language translation."""
blob=TextBlob(text)
sent=blob.sentences
words=blob.words
noun_phrases=blob.noun_phrases
print('Sentences:')
for s in sent:
    print('-',s)
print('\nWords:')
print(words)

print('\nNoun Phrases')
print(noun_phrases)
```

```python
from textblob import TextBlob
import nltk

# Ensure necessary NLTK data is downloaded
nltk.download('punkt')

# Step 1: Paste the article's text here
text = """
India has announced measures targeting Pakistan, a day after 26 people were killed by gunmen at a Himalayan tourist attraction in Indian-administered Kashmir.

They include the closure of the main border crossing linking the two countries, the suspension of a landmark water-sharing treaty, the expulsion of diplomats and an order for some Pakistani visa holders to leave within 48 hours.

Pakistani officials, who have denied the country's role in the attack, are meeting on Thursday to come up with a response, its foreign minister said.

The attack in Pahalgam was one of the deadliest in recent memory for the restive region - and is threatening to aggravate already icy bilateral ties.
"""

# Step 2: Create a TextBlob object
blob = TextBlob(text)

# Step 3: Display overall sentiment of the article
print(" Overall Article Sentiment:")
print("Polarity:", blob.sentiment.polarity)
print("Subjectivity:", blob.sentiment.subjectivity)

# Step 4: Display sentiment for each sentence
print("\n Sentiment by Sentence:")
for i, sentence in enumerate(blob.sentences):
    print(f"\nSentence {i+1}:")
    print(sentence)
    print("Polarity:", sentence.sentiment.polarity)
    print("Subjectivity:", sentence.sentiment.subjectivity)
```
